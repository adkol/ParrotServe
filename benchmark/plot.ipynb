```python
import re
import matplotlib
import matplotlib.pyplot as plt
from collections import defaultdict
import numpy as np

WRYH = matplotlib.font_manager.FontProperties(fname = '/home/v-chaofanlin/fonts/msyh.ttf')
```

```python
# Draw 5.2 figure 1

# bs: 16, 32, 64, 128
# seq_len: 1024, 2048, 4096, 8192

bs = [16, 32, 64, 128]
bs_data_vllm = [52.20, 97.20, 214.33, 472.24]
bs_data_srt = [30.28, 30.33, 35.64, 53.41]

systems = ["PagedAttention", "SharedAttention"]
hatches = ["", "\\"]
colors = ["#fc8d59", "#4575b4"]

# Generate the chart
x = np.arange(len(bs))
width = 0.25
fig, ax = plt.subplots()
plt.grid(True)

for i, system in enumerate(systems):
    if system == "PagedAttention":
        data = bs_data_vllm
    else:
        data = bs_data_srt

    rects = ax.bar(
        x - width / 2 + i * width,
        data,
        width,
        hatch=hatches[i],
        color=colors[i],
        label=system,
        zorder=3,
    )  # hatches

plt.legend(
    loc="upper left",
    prop={"size": 15},
)

ax.tick_params(axis="y", labelsize=20, direction="in")
ax.tick_params(axis="x", labelsize=20, direction="in")
ax.set_xlabel(u"批大小", fontsize=26, fontproperties = WRYH)
ax.set_ylabel(u"单次解码时延 (ms)", fontsize=26, fontproperties = WRYH)
ax.set_xticks([_ + 0.1 for _ in x])
ax.set_xticklabels(bs)
plt.ylim([0, 500])

fig.tight_layout()

plt.show()
```
